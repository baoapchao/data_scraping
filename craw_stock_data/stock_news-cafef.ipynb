```python
# define params

news_categories = [{'id' : 2, 'category' : 'Trả cổ tức - Chốt quyền'}
                    , {'id' : 1 , 'category' : 'Tình hình SXKD & Phân tích khác'}
                    , {'id' : 4 , 'category' : 'Tăng vốn - Cổ phiếu quỹ'}
                    , {'id' : 5 , 'category' : 'GD cổ đông lớn & Cổ đông nội bộ'}
                    , {'id' : 3 , 'category' : 'Thay đổi nhân sự'}
                    ]

request_url = fr"https://s.cafef.vn/Ajax/Events_RelatedNews_New.aspx"

list_year = range(2017, 2023)

# start to code
for symbol in symbols:
    lst = []
    for category in news_categories:

        prmts = {'symbol' : symbol , 'floorID' :  0, 'configID' : category['id'] , 'PageIndex' : 1 , 'PageSize' : 10000, 'Type' : 2}
        r = requests.get(request_url, params = prmts)

        try: 
            soup = bs(r.text, 'html.parser')

            a = soup.find('div', id = 'divEvents')
            list_events = a.find_all('li')
            for event in list_events:
                event_name = event.find('a')
                event_time = event.find('span')
                lst.append({'event_name' : event_name.text , 'event_time' : event_time.text , 'category' : category['category']})
        except:
            print(f"Not found {category['category']} for {symbol}")
    if len(lst) > 0:
        df = pd.DataFrame(lst)
        df['symbol'] = symbol
        for year in list_year:
            df_filtered = filter_datetime(df, 'event_time', '%d/%m/%Y %H:%M', 'year', year)
            df_to_csv(df_filtered, 'stock_news', 'cafef', year, symbol)
```
