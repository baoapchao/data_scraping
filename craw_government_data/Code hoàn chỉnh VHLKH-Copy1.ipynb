```python
from bs4 import BeautifulSoup as bs
import urllib
import lxml
import pandas as pd
import time
```

```python
detai_form_url = 'https://vast.gov.vn/web/guest/cac-e-tai-nckh-ptcn?p_p_id=showview_WAR_VastCapGiayChungNhanDeTaiportlet&p_p_lifecycle=0&p_p_state=normal&p_p_mode=view&p_p_col_id=column-1&p_p_col_count=1&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_mvcPath=%2Fhtml%2Fshowview%2Fview.jsp&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_delta=20&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_keywords=&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_advancedSearch=false&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_andOperator=true&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_resetCur=false&_showview_WAR_VastCapGiayChungNhanDeTaiportlet_cur='
listk = range(1,75)
dict1 = {}

#lay list tung trang page de tai
for k in listk:
    detai_urls.append(detai_form_url + str(k))
    
detai_chitiet_urls = []
#tao function lay url cac trang chi tiet:  
def get_detai_chitiet_urls(detai_urls):
    for detai_url in detai_urls:
        soup_all_page = bs(urllib.request.urlopen(detai_url), 'lxml')
        elements = soup_all_page.find_all('td',class_="table-cell w40")
        detai_chitiet_urls.extend([element.find('a')['href'] for element in elements])
```

```python
# chay code lay tat ca url cac trang chi tiet tung de tai:
start = time.time()
get_detai_chitiet_urls(detai_urls)
end = time.time()
print(f"Runtime of the program is {end - start}")
```

```python
print(detai_chitiet_urls)
```

```python
#tao function lay columns:        
def get_key_list(detai_chitiet_urls):
    global keys
    keys = []
    for detai_chitiet_url in detai_chitiet_urls:
        soupchitiet = bs(urllib.request.urlopen(detai_chitiet_url), 'lxml')
        lines = soupchitiet.find_all('tr')
        
        #vong lap tung dong cua trang chi tiet
        for line in lines:
            try:
                title = line.find('td', class_= 'detai-title')
                keys.append(title.text.replace('\n','').replace('\t',""))
            except:
                pass
```

```python
start = time.time()
get_key_list(detai_chitiet_urls)
end = time.time()
print(f"Runtime of the program is {end - start}")
```

```python
# in thu cac key
for key in keys:
    print(key)
```

```python
global dict1
dict1 = dict.fromkeys(keys)
keys = dict1.keys()
for key in keys:
    dict1[key] = []
print(dict1)
print(keys)
```

```python
#tao function fill vao dictionary du lieu
def fill_vao_dict(detai_chitiet_urls):
    #vong lap tung trang chi tiet
    for detai_chitiet_url in detai_chitiet_urls:
        soupchitiet = bs(urllib.request.urlopen(detai_chitiet_url), 'lxml')
        lines = soupchitiet.find_all('tr')
        list_available_title = []
        
        #vong lap tung dong cua trang chi tiet
        for line in lines:
            try:
                #tim title
                title = line.find('td', class_= 'detai-title')
                titlename = title.text.replace('\n','').replace('\t',"").replace('\xa0',"")
                
                #tim content va fill vao dict1
                content = line.find('td', class_='detai-content')
                contentname = content.text.replace('\n','').replace('\t',"").replace('\xa0',"")
                dict1[titlename].append(contentname)
                
                #tim tat ca cac title co trong trang chi tiet
                list_available_title.append(titlename)
            except:
                pass 
        
        #neu key ko co trong list cac title trong trang chi tiet thi fill blank vao dict1
        for key in keys:
            if key not in list_available_title:
                dict1[key].append('')
            else:
                pass
```

```python
# chay code de fill vao dictionary du lieu
start = time.time()
fill_vao_dict(detai_chitiet_urls)
end = time.time()
print(f"Runtime of the program is {end - start}")
```
