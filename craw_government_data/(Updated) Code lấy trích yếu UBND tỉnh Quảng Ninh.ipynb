```python
import selenium as sl
from selenium import webdriver
import pandas as pd
import numpy as np
import csv
import time
from bs4 import BeautifulSoup as bs
```

```python
Webchidao = "https://www.quangninh.gov.vn/Trang/van-ban-chi-dao.aspx"
driver = webdriver.Chrome(executable_path= r"C:\Users\ADMIN\OneDrive\__Study\Python\chromedriver.exe")
driver.get(Webchidao)
```

```python
# old fashioned
k = 1

so = []
ngay = []
trichyeu = []
coquan = []
dinhkem = []

tapdata = []

while k>0 and k<50:
    datas = driver.find_elements_by_class_name("textleft")
    
    for data in datas:
        if k%5 == 0:
            tapdata.append(data.text)
            so.append(tapdata[0])
            ngay.append(tapdata[1])
            trichyeu.append(tapdata[2])
            coquan.append(tapdata[3])
            dinhkem.append(tapdata[4])
            tapdata = []
        else:
            tapdata.append(data.text)
        k = k + 1
    try:
        button_next = driver.find_element_by_css_selector("#ctl00_ctl39_g_4c835896_8e18_4b5b_a775_babbc01541e9_ctl00_btnNext")
        button_next.click()
    except:
        break
```

```python
# do this instead 
df = pd.DataFrame([])
while True:
    soup = bs(driver.page_source, 'lxml')
    tables = pd.read_html(soup.prettify())
    if df.empty:
        df = tables[1]
        print(df)
        
    else:
        df = df.append(tables[1])
    try:
        button_next = driver.find_element_by_css_selector("#ctl00_ctl39_g_4c835896_8e18_4b5b_a775_babbc01541e9_ctl00_btnNext")
        button_next.click()
    except:
        break
```
